# Multiple Linear Regression

## Definition

Multiple Linear Regression (MLR) models the relationship between one dependent variable and two or more independent variables.

It extends simple linear regression to higher dimensions.

## 2. Mathematical Model

𝑦=𝛽0+𝛽1𝑥1+𝛽2𝑥2+⋯+𝛽𝑛𝑥𝑛+𝜖
Where:
- y → Dependent (target) variable
- x1 , x2 , x3 ...xn → Independent variables
- β0 → Intercept
- 𝛽0+𝛽1,...𝛽n → Coefficients
- ϵ → Error term

## 3. Key Idea
- The model fits a hyperplane (not a line) in multi-dimensional space.
- Each feature contributes linearly to the output.

```python

from sklearn.datasets import make_regression
import pandas as pd
import numpy as np

import plotly.express as px
import plotly.graph_objects as go

from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
X,y = make_regression(n_samples=100, n_features=2, n_informative=2, n_targets=1, noise=50)
```

```python
df = pd.DataFrame({'feature1':X[:,0],'feature2':X[:,1],'target':y})
```

```python
df.shape
```

```python

df.head()
```

```python
fig = px.scatter_3d(df, x='feature1', y='feature2', z='target')

fig.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
print("MAE",mean_absolute_error(y_test,y_pred))
print("MSE",mean_squared_error(y_test,y_pred))
print("R2 score",r2_score(y_test,y_pred))
```

```python
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)

xGrid, yGrid = np.meshgrid(x, y)

# Create input features (100 samples, 2 features)
final = np.vstack((xGrid.ravel(), yGrid.ravel())).T

# Predict
z_final = lr.predict(final)

# Reshape to grid
z = z_final.reshape(10, 10)
```
