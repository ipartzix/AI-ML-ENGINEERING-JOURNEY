# Multiple Linear Regression

## Definition

Multiple Linear Regression (MLR) models the relationship between one dependent variable and two or more independent variables.

It extends simple linear regression to higher dimensions.

## Mathematical Model

𝑦=𝛽0+𝛽1𝑥1+𝛽2𝑥2+⋯+𝛽𝑛𝑥𝑛+𝜖
Where:
- y → Dependent (target) variable
- x1 , x2 , x3 ...xn → Independent variables
- β0 → Intercept
- 𝛽0+𝛽1,...𝛽n → Coefficients
- ϵ → Error term

## Key Idea
- The model fits a hyperplane (not a line) in multi-dimensional space.
- Each feature contributes linearly to the output.

## Example

Predict house price using:

- Area (𝑥1)
- Number of rooms (𝑥2)
- Location score (𝑥3)
- Model:        
       ```  Price=β0​+β1​(Area)+β2​(Rooms)+β3​(Location)```

## Assumptions

MLR works properly only if these conditions hold:

- Linearity:-
Relationship between input and output is linear
- Independence of errors:-
Residuals are not correlated
- Homoscedasticity:-
Constant variance of errors
- No Multicollinearity:-
Independent variables should not be highly correlated
- Normality of residuals:-
Errors should follow normal distribution

## Cost Function (Loss Function)
J(β)=n1​∑i=1n​(yi​−y^​i​)2
- This is Mean Squared Error (MSE)
- Goal: Minimize error between actual and predicted values

## Methods to Find Coefficients
- Normal Equation
- Gradient Descent (used for large datasets)
- OLS (Ordinary Least Squares)

## Evaluation Metrics

Used to measure model performance:

- R² Score (Coefficient of Determination)
- Adjusted R²
- Mean Absolute Error (MAE)
- Mean Squared Error (MSE)
- Root Mean Squared Error (RMSE)

## Advantages
- Simple and interpretable
- Works well for linear relationships
- Fast to train

##  Disadvantages
- Cannot capture non-linear patterns
- Sensitive to outliers
- Multicollinearity can break the model
- Assumptions must hold

````python

from sklearn.datasets import make_regression
import pandas as pd
import numpy as np

import plotly.express as px
import plotly.graph_objects as go

from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
````

````python
X,y = make_regression(n_samples=100, n_features=2, n_informative=2, n_targets=1, noise=50)
````

````python
df = pd.DataFrame({'feature1':X[:,0],'feature2':X[:,1],'target':y})
````

````python
df.shape
````

````python

df.head()
````

````python
fig = px.scatter_3d(df, x='feature1', y='feature2', z='target')

fig.show()
````

````python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=3)
````

````python
from sklearn.linear_model import LinearRegression
````

````python
lr = LinearRegression()
````

````python
lr.fit(X_train,y_train)
````

````python
y_pred = lr.predict(X_test)
````

````python
print("MAE",mean_absolute_error(y_test,y_pred))
print("MSE",mean_squared_error(y_test,y_pred))
print("R2 score",r2_score(y_test,y_pred))
````
