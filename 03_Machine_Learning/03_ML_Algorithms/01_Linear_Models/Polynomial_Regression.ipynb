## Polynomial Regression – Machine Learning Notes

### 1. Overview
Polynomial Regression is an extension of Linear Regression that models the relationship between the independent variable `X` and dependent variable `y` as an **nth-degree polynomial**.

It is useful when data shows a **non-linear trend** but can still be expressed as a polynomial function.

---
### 2. Mathematical Representation

For degree `n`:

y = b0 + b1x + b2x² + b3x³ + ... + bnxⁿ

Where:
- `b0, b1, ..., bn` → coefficients (parameters)
- `x` → input feature
- `y` → predicted output

---
### 3. Key Idea

Although the equation looks non-linear, **Polynomial Regression is still a linear model** because:
- It is linear in terms of coefficients (`b0, b1, b2, ...`)

---
### 4. When to Use

Use Polynomial Regression when:
- Data follows a curve (not a straight line)
- Linear Regression underfits the data
- Relationship between variables is non-linear

---
### 5. Example

Linear model (underfitting):
y = b0 + b1x

Polynomial model:
y = b0 + b1x + b2x²

---
### 6. Steps to Implement

1. Import libraries
2. Load dataset
3. Transform features into polynomial features
4. Train model using Linear Regression
5. Predict and evaluate

---

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression,SGDRegressor

from sklearn.preprocessing import PolynomialFeatures,StandardScaler

from sklearn.metrics import r2_score

from sklearn.pipeline import Pipeline
```

```python
X = 6 * np.random.rand(200, 1) - 3
y = 0.8 * X**2 + 0.9 * X + 2 + np.random.randn(200, 1)

# y = 0.8x^2 + 0.9x + 2
```

```python
plt.plot(X, y,'b.')
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
# Train test split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
# Applying linear regression
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
r2_score(y_test,y_pred)
```

```python
plt.plot(X_train,lr.predict(X_train),color='r')
plt.plot(X, y, "b.")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
# Applying Polynomial Linear Regression
# degree 2
poly = PolynomialFeatures(degree=2,include_bias=True)

X_train_trans = poly.fit_transform(X_train)
X_test_trans = poly.transform(X_test)
```

```python
print(X_train[0])
print(X_train_trans[0])
```

```python
# include_bias parameter
```

```python
lr = LinearRegression()
lr.fit(X_train_trans,y_train)
```

```python
y_pred = lr.predict(X_test_trans)
```

```python
r2_score(y_test,y_pred)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python

X_new=np.linspace(-3, 3, 200).reshape(200, 1)
X_new_poly = poly.transform(X_new)
y_new = lr.predict(X_new_poly)
```

```python
plt.plot(X_new, y_new, "r-", linewidth=2, label="Predictions")
plt.plot(X_train, y_train, "b.",label='Training points')
plt.plot(X_test, y_test, "g.",label='Testing points')
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
