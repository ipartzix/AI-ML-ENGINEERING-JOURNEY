```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/refs/heads/main/day48-simple-linear-regression/placement.csv")
```

```python
df.head()
```

```python
plt.scatter(df["cgpa"],df["package"])
plt.xlabel("CGPA")
plt.ylabel("PackAge(in lpa)")
```

```python
X = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
X
y
```

```python
from sklearn.model_selection import train_test_split
X_train ,X_test ,y_train , y_test = train_test_split(X,y , random_state=42,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
X_test
```

```python
y_test
```

```python
lr.predict(X_test.iloc[0].values.reshape(1,1))
lr.predict(X_test.iloc[5].values.reshape(1,1))
```

```python
plt.scatter(df["cgpa"],df["package"])
plt.plot(X_train,lr.predict(X_train),color ="red")
plt.xlabel("CGPA")
plt.ylabel("PackAge(in lpa)")
```

```python
m = lr.coef_
```

```python
b = lr.intercept_
```

## Advantages
Simple and easy to understand
Fast computation
Works well with linear data

## Limitations
Cannot capture non-linear relationships
Sensitive to outliers
Assumes strict conditions (linearity, normality)

## Real-world Examples
Predicting house prices (based on size)
Salary prediction (based on experience)
Sales forecasting

## Visualization Idea
Scatter plot of data points
Best-fit straight line passing through points

## Key Takeaway

Simple Linear Regression tries to find the best straight line that explains the relationship between input and output by minimizing error.
