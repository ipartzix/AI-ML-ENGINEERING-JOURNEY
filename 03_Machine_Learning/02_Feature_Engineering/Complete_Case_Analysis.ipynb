# Complete Case Analysis

## 1. Definition
**Complete Case Analysis (CCA)** means:

> Only those rows (observations) that have **no missing values at all** are kept.  
> All rows with **any missing value** are removed.

---

## 2. Example

### Original Dataset

| Name | Age | Salary |
|------|-----|--------|
| A    | 25  | 50000  |
| B    | NaN | 60000  |
| C    | 30  | NaN    |
| D    | 28  | 55000  |

### After CCA

| Name | Age | Salary |
|------|-----|--------|
| A    | 25  | 50000  |
| D    | 28  | 55000  |

Rows **B** and **C** are removed because they contain missing values.

---
## 3. When to Use

- Missing data is **very small** (e.g., <5%)
- Data is **Missing Completely At Random (MCAR)**
- You want a **simple and fast method**

---

## 4. Advantages

- Simple to implement  
- No need for complex imputation  
- Keeps dataset consistent (no artificial values)  
- Works well for clean, small datasets  

---

## 5. Disadvantages

- Loss of data  
- Can introduce **bias** if data is not MCAR  
- Reduces statistical power  
- Not suitable for datasets with many missing values  

---

## 6. Assumption (Important)

CCA assumes:

- Missing values are **Missing Completely At Random (MCAR)**

If this assumption is violated, results can be misleading.
