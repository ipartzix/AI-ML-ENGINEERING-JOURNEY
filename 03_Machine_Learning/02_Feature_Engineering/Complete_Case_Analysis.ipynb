# Complete Case Analysis

## 1. Definition
**Complete Case Analysis (CCA)** means:

> Only those rows (observations) that have **no missing values at all** are kept.  
> All rows with **any missing value** are removed.

---

## 2. Example

### Original Dataset

| Name | Age | Salary |
|------|-----|--------|
| A    | 25  | 50000  |
| B    | NaN | 60000  |
| C    | 30  | NaN    |
| D    | 28  | 55000  |

### After CCA

| Name | Age | Salary |
|------|-----|--------|
| A    | 25  | 50000  |
| D    | 28  | 55000  |

Rows **B** and **C** are removed because they contain missing values.
