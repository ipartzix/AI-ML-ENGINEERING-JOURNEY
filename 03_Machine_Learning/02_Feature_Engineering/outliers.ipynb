# Introduction to Outliers

Outliers are data points that are significantly different from the majority of observations in a dataset. They can distort statistical analyses and affect machine learning models if not handled properly.

## Key Characteristics
- **Extreme Values:** Outliers are unusually high or low compared to other data points.
- **Impact:** Can affect mean, variance, correlation, and regression analysis.
- **Types:**
  1. **Point Outliers:** Single data point deviates from the rest.
  2. **Contextual Outliers:** Abnormal only in a specific context.
  3. **Collective Outliers:** A group of points collectively deviates from the pattern.

## Causes of Outliers
- Measurement errors or data entry mistakes
- Natural variability in the data
- Experimental or sampling anomalies

## Detection Methods
- **Statistical Methods:**
  - Z-score (\|Z\| > 3)
  - Interquartile Range (IQR)
- **Visual Methods:**
  - Box plot
  - Scatter plot
- **Machine Learning Methods:**
  - Isolation Forest
  - DBSCAN clustering
  - One-Class SVM

## Handling Outliers
- **Remove:** If caused by error
- **Transform:** Apply log or square root transformations
- **Cap/Floor:** Limit extreme values
- **Keep:** If they provide meaningful insight

> **Note:** Outliers should be carefully analyzed before deciding how to handle them, as they can contain valuable information about anomalies or rare events.
