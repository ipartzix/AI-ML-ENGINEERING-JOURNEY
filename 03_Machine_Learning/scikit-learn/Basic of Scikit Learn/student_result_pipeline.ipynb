### Step 3 -> Encode the columns  

```python
preprocessor = ColumnTransformer([
    ('num', SimpleImputer(strategy='mean'),
    ['Attendance','PreviousMarks','StudyHours']),
    
    ('cat', OneHotEncoder(drop='first'),
    ['Participation'])
])
```

###  Numerical Imputation using ColumnTransformer

This step handles **missing values in numerical features**.

####  What is `num_impute`?

* `num_impute` is just a **name (label)** given to this transformation step.
* It stands for:

  * **num** → numerical
  * **impute** → fill missing values

 So, it means: *“fill missing values in numerical columns”*

---

####  What does this transformation do?

* Applies **mean imputation** on:

  * Attendance
  * PreviousMarks
  * StudyHours

* If a value is missing → it is replaced with the **average of that column**

* If no value is missing → it remains unchanged

---

####  What is `remainder='passthrough'`?

* Keeps all **other columns unchanged**
* Example:

  * `Participation` is not modified here
  * It will pass through as it is

---

####  Summary

* Numerical columns → missing values filled using mean
* Other columns → left unchanged
* `num_impute` → just a readable name for this step

---

####  Purpose
To ensure the dataset has **no missing numerical values** before training the machine learning model.

## Prepare pipelines

```python
# Predict
y_pred = pipe.predict(X_test)
```
