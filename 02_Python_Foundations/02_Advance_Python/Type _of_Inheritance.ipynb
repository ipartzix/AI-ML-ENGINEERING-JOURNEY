# Level of Inheritance

## 1. Single Level Inheritance
One parent → one child

```python
class Phone:
    def call(self):
        print("Calling...")


class SmartPhone(Phone):
    def internet(self):
        print("Using Internet")


s = SmartPhone()
s.call()
s.internet()
```

## 2. Multilevel Inheritance
Parent → Child → Grandchild

```python
class Phone:
    def call(self):
        print("Calling...")


class SmartPhone(Phone):
    def internet(self):
        print("Using Internet")


class AndroidPhone(SmartPhone):
    def play_store(self):
        print("Opening Play Store")


a = AndroidPhone()
a.call()
a.internet()
a.play_store()
```

## 3. Hierarchical Inheritance
One parent → multiple children

```python
class Phone:
    def call(self):
        print("Calling...")


class AndroidPhone(Phone):
    def android_feature(self):
        print("Android features")


class IPhone(Phone):
    def ios_feature(self):
        print("iOS features")


a = AndroidPhone()
i = IPhone()

a.call()
a.android_feature()

i.call()
i.ios_feature()
```

## 4. Multiple Inheritance
Multiple parents → one child

```python
class Camera:
    def photo(self):
        print("Taking photo")


class Music:
    def play_music(self):
        print("Playing music")


class SmartPhone(Camera, Music):
    pass


s = SmartPhone()
s.photo()
s.play_music()
```

## 5. Hybrid Inheritance
Combination of multiple + multilevel

```python
class Phone:
    def call(self):
        print("Calling...")


class Camera(Phone):
    def photo(self):
        print("Taking photo")


class Music:
    def play_music(self):
        print("Playing music")


class SmartPhone(Camera, Music):
    def internet(self):
        print("Using Internet")


s = SmartPhone()
s.call()
s.photo()
s.play_music()
s.internet()
```

## summary

### Type	                     Structure
### Single	                      A → B
### Multilevel	                  A → B → C
### Hierarchical	              A → B, A → C
### Multiple	                  A + B → C
### Hybrid	                      Mix of two or more
